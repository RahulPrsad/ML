```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Height (in feet) and Weight (in kg)
x= np.array([5.0,5.1,5.2,5.3,5.4,5.5,5.6,5.7,5.8])
y = np.array([53,55,59,61,65,68,70,74,76])
n=len(x)
```

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x2 = np.sum(x * x)
sum_xy = np.sum(x * y)

# Step 2: least squares formulas
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
a = (sum_y / n) - b * (sum_x / n)

print("Intercept (a):", a)
print("Slope (b):", b)

x_new = 5.9
y_pred = a + b * x_new

print("Predicted weight for height 5.9:", y_pred)

y_hat = a + b * x

mse = np.mean((y - y_hat)**2)
print("MSE:", mse)
mae = np.mean(np.abs(y - y_hat))
print("MAE:",mae)
```
