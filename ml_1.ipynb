```python
import csv
def find_s(traing_set):

  hypothsis=['$']*(len(traing_set[0])-1)

  step=0
  for ex in traing_set:
    step+=1
    print(f"Step {step}")
    print(hypothsis)
    if ex[-1]=='Yes':
      for i in range(len(ex)-1):
        if(hypothsis[i]=='$'):
          hypothsis[i]=ex[i]
        elif(hypothsis[i]!=ex[i]):
          hypothsis[i]='?'


  return hypothsis

with open('/content/sample_data/ws.csv') as f:
  data = list(csv.reader(f))
  print("Main Data:")
  for row in data:
    print(row)


find_s(data)
```
