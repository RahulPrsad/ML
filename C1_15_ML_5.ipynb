```python
import numpy as np
X = np.array([
    [5.0, 45],
    [5.11, 26],
    [5.6, 30],
    [5.9, 34],
    [4.8, 40],
    [5.8, 36],
    [5.3, 19],
    [5.8, 28],
    [5.5, 23],
    [5.6, 32]
])
y = np.array([77, 47, 55, 59, 72, 60, 40, 60, 45, 58])

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))
x_check = np.array([5.5, 38])

```

```python
def knn_reg(X,y,x,k):
  distance=[]
  for i in range(0,len(X)):
    distance.append((euclidean_distance(x,X[i]), y[i]))
  distance.sort(key=lambda x:x[0])
  distance=distance[:k]
  prediction=np.mean([ dist[1]for dist in distance])
  return prediction
```

```python
def weight_knn_reg(X,y,x,k):
  distance=[]
  for i in range(0,len(X)):
    distance.append((euclidean_distance(x,X[i]), y[i]))
  distance.sort(key=lambda x:x[0])
  distance=distance[:k]
  numerator = 0
  denominator = 0
  for dist, value in distance:
      weight = 1 / (dist + 1e-5)
      numerator += weight * value
      denominator += weight
  prediction = numerator / denominator
  return prediction
```

```python
# Simple KNN
print("Simple KNN (k=3):", knn_reg(X, y, x_check, 3))
print("Simple KNN (k=5):", knn_reg(X, y, x_check, 5))

# Weighted KNN
print("Weighted KNN (k=3):", weight_knn_reg(X, y, x_check, 3))
print("Weighted KNN (k=5):", weight_knn_reg(X, y, x_check, 5))
```
